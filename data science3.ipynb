```python
import pandas as pd
def convert_price_cell(cell):
    if cell=="n.a.":
        return '78'
    return cell
def convert_oops_cell(cell):
    if cell=="not available":
        return 'NaN'
    return cell
df=pd.read_excel('C:\\Users\\USER\\Desktop\\SASA\\testing data.xlsx',converters={
    'price':convert_price_cell,
    'oops':convert_oops_cell
})
df.head()
```

```python
df.to_excel("new.xlsx",sheet_name="stock",startrow=1,startcol=2)
```

```python
import pandas as pd
df_stocks=pd.DataFrame({
    'tickets':['google','wmst','mstc'],
    'price':[78,65,50],
    'pe':[30.67,78.90,56.65],
    'oops':[27.82,4.61,5.6]
})
df_weather=pd.DataFrame({
    'day':['1/1/2017','1/2/2017','1/3/2017'],
    'temperature':[32,35,28],
    'event':['rain','sunny','snow']
})
with pd.ExcelWriter('stocks_weather.xlsx')as writer:
    df_stocks.to_excel(writer,sheet_name="dual")
    df_weather.to_excel(writer,sheet_name="weather")
```
